**Aquisição dos Dados**

```python
import urllib.request  
import nltk # tentativa de tirar os stopwords 
nltk.download('stopwords')
from nltk.corpus import stopwords

SW = stopwords.words('portuguese') # tentativa de tirar os stopwords

texto = []

for line in urllib.request.urlopen('https://raw.githubusercontent.com/NatanaelSSilva/Arquivos-Mackenzie/master/Artigo.Inteligencia.Artificial.txt'): 
    texto.append(line.decode('utf-8'))
# transformar o texto 

for i in range(len(texto)):
    texto[i] = texto[i].lower() # transformar o texto em minuscula 
    texto[i] = texto[i].replace('\n','')
    texto[i] = texto[i].replace('.','')
    texto[i] = texto[i].replace(',','')
    texto[i] = texto[i].replace('(','')
    texto[i] = texto[i].replace(')','')
    texto[i] = texto[i].replace('?','')
    texto[i] = texto[i].replace('\'','') # substitui os itens da esquerda por vazio " "

```

**Construção do Dicionário**

```python
mydict = {} # criar um dicionário
                            
for line in texto:

  line = line.lower()                  
  words = line.split()  # separa cada palavra

  for word in words:                      
    if word not in mydict.keys():         # se palavra não está no dicionário 
      mydict[word] = 1                    # acrescenta a palavra com o valor 1
    else:                                 # se a entrada já existe 
      mydict[word] =  mydict[word] + 1    # apenas soma 1 ao valor já existente

print(mydict) # imprimir dicionário
```

**Exibição dos Resultados**

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl # para gráficos
import matplotlib.pyplot as plt

df = pd.DataFrame(mydict.items(),  columns=['word', 'count']).sort_values('count',ascending=False)
df = df[df['count'] > 4] # somente termos com mais de 4 ocorrências

plt.figure(figsize=(24,10))
mpl.style.use(['seaborn'])
sns.barplot(x=df.word,y=df['count']) # gráfico de barras
plt.xticks(rotation=90) 

plt.show()
```
